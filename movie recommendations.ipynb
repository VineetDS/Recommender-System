```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies
```

```python
credits
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies
```

```python
movies = movies[['genres','keywords','movie_id','title','cast','crew','overview']]
```

```python
movies
```

```python
movies.dropna(inplace= True)
```

```python
movies.isna().sum()
```

```python
import ast
def my_fun(x):
    L = []
    for i in ast.literal_eval(x):
        L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(my_fun)
```

```python
movies['keywords'] = movies['keywords'].apply(my_fun)
```

```python
def my_fun_new(x):
    L = []
    counter = 0
    for i in ast.literal_eval(x):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(my_fun_new)
```

```python
def new_fun(x):
    L = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
        
```

```python
movies['crew'] = movies['crew'].apply(new_fun)
```

```python
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies = movies[['movie_id','title','tag']]
```

```python
movies['tag'] = movies['tag'].apply(lambda x:" ".join(x))
```

```python
movies
```

```python
movies['tag'] = movies['tag'].apply(lambda x :x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000, stop_words= 'english')
```

```python
cv.fit_transform(movies['tag']).toarray().shape
```

```python
vectors = cv.fit_transform(movies['tag']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer                  # used to avoid repetion on words by root word
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tag'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity[1]
```

```python
sorted(similarity[0],reverse = True)
```

```python
enumerate(similarity[0])                     # enumerate used here to show the relation of index position
```

```python
sorted(list(enumerate(similarity[0])),reverse= True,key = lambda x:x[1])[1:6]
```

```python
movies[movies['title'] == 'Avatar'].index[0]
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse= True,key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
